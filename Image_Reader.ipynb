```python
from google.colab import userdata
GEMNI_API_KEY = userdata.get('GOOGLE_API_KEY')
```

```python
import google.generativeai as genai
```

```python
genai.configure(api_key=GEMNI_API_KEY)
```

```python
model = genai.GenerativeModel("gemini-1.5-flash")
```

```python
import requests
from PIL import Image
from io import BytesIO
from IPython.display import display

# Here is the online image url
url = 'https://www.simplilearn.com/ice9/free_resources_article_thumb/Why_E-Learning_Insights_into_the_World_of_Online_Learning_and_Development.jpg'

response = requests.get(url)

img = Image.open(BytesIO(response.content))

display(img)
```

```python
prompt = """ Describe the image  in detail """


response = model.generate_content([prompt,img])
print(response.text)
```
